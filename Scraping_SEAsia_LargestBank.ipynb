```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_banks_in_Southeast_Asia'

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table', class_='wikitable sortable')
```

```python
table = soup.find_all('table')[0]
print(table)
```

```python
comp_titles = table.find_all('th')
```

```python
comp_titles
```

```python
comp_table_titles = [title.text.strip() for title in comp_titles]
print(comp_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=comp_table_titles)
```

```python
df
```

```python
# in wikipedia page inspect <tr> means row and <td> means data
column_data = table.find_all('tr')
```

```python
for row in column_data[1:] :
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
